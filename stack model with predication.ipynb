```python
data = pd.read_excel('p.values.xlsx', sheet_name = 'Totaldata (wrapper-filter) (3)', header = None).transpose()
labels = pd.read_excel('p.values.xlsx',sheet_name = 'Label', header = None).transpose()


data.head()
```

```python


def stack_models():
    
    estimators = [
    ('dtc', DecisionTreeClassifier(random_state=42)), 
    ('rfc', RandomForestClassifier(n_estimators=50, random_state=42)),
    ('gbdt', GradientBoostingClassifier(random_state = 42)),
    ('knn', KNeighborsClassifier())]
    
    classifier = StackingClassifier(estimators=estimators, final_estimator = LogisticRegression(random_state=42))
    

    return classifier

classifier = stack_models()

features_train, features_test, target_train, target_test = train_test_split(

data, labels, test_size=0.3, random_state=1)
 

classifier.fit(features_train, target_train)

lr_model = LogisticRegression(random_state=42)
lr_model.fit(classifier.transform(features_train), target_train)



kf = KFold(n_splits=10, shuffle=True, random_state=1)
 

cv_results = cross_val_score(classifier,

data, 

labels, 

cv=kf, 

scoring="accuracy", 

n_jobs=-1) 
 

cv_results.mean()




importance = classifier.named_estimators_['rfc'].feature_importances_
n_features = len(importance)
importance = classifier.named_estimators_['rfc'].feature_importances_
n_features = len(importance)
feature_names = ['Feature1', 'Feature2', 'Feature3', 'Feature4','Feature5'] 

total_importance = np.sum(importance)

feature_importance_percentages = [(importance[i] / total_importance) * 100 for i in range(n_features)]

for i in range(n_features):
    print(f'{feature_names[i]}: {feature_importance_percentages[i]:.2f}%')


 
# Predict values for training target vector

target_hat = classifier.fit(features_train, target_train).predict(features_test)
 
# Calculate accuracy

accuracy_score(target_test, target_hat)








 
```
